```python
import csv
import pandas as pd
from transformers import BertTokenizer, BertForSequenceClassification
import torch

# Initialize the tokenizer and model
model_path = 'aubmindlab/bert-base-arabertv02'
tokenizer = BertTokenizer.from_pretrained(model_path, do_lower_case=True)
model = BertForSequenceClassification.from_pretrained(model_path, num_labels=2)  # assuming 2 classes

def combine_context_glosses(context, glosses):
    # Combine context with each gloss
    combined_pairs = []
    for gloss in glosses:
        # Check if gloss has the expected format
        if "@" not in gloss:
            continue

        # Gloss ID starts with 302 for Ghani and 303 for Modern lexicon
        lexicon_source = 'Ghani' if gloss.startswith('302') else 'Modern'

        # Split the gloss into concept ID and percentage
        concept_id, percentage = gloss.split('@', 1)  # Limit split to only 1 occurrence

        # Combine context and gloss as a pair
        pair = {
            'context': context,
            'lexicon_source': lexicon_source,
            'concept_id': concept_id,
            'percentage': percentage
        }
        combined_pairs.append(pair)

    return combined_pairs

def perform_inference(sentence, glosses):
    # Combine context and glosses
    pairs = combine_context_glosses(sentence, glosses)

    # Perform inference for each pair
    logits_list = []
    for pair in pairs:
        # Create input tensors for the pair
        inputs = tokenizer(pair['context'], pair['concept_id'], return_tensors='pt', padding=True, truncation=True)

        # Perform inference
        with torch.no_grad():
            outputs = model(**inputs)

        logits = outputs.logits
        logits_list.append(logits)

    return logits_list, pairs

# Read sentences from CSV
df1 = pd.read_csv('/content/drive/MyDrive/WSD_corpus_context.csv', delimiter=';', quoting=csv.QUOTE_NONE)
sentences = df1.iloc[:, 0].tolist()

# Read glosses from CSV
glosses = []
with open('/content/drive/MyDrive/lemmas.csv', 'r', encoding='utf-8') as csvfile:
    reader = csv.reader(csvfile)
    for row in reader:
        if len(row) < 1:
            continue
        glosses.append(row[0])

# Initialize statistics counters
top_model_rank = [0, 0, 0]  # [Ranked 1st, Ranked 2nd, Ranked 3rd]
top_user_rank = [0, 0, 0]   # [Ranked 1st, Ranked 2nd, Ranked 3rd]
total_lemmas = 0

# Process each sentence
for sentence in sentences:
    # Perform inference for the sentence
    logits_list, pairs = perform_inference(sentence, glosses)

    # Process each pair
    for idx, (logits, pair) in enumerate(zip(logits_list, pairs)):
        # Sort logits to get the model's rank order
        sorted_indices = torch.argsort(logits, descending=True).squeeze().tolist()
        model_rank_order = [sorted_indices.index(i) + 1 for i in range(len(sorted_indices))]

        # Get the user's rank order
        user_rank_order = int(pair['percentage'].split(':')[0])

        # Update statistics
        if user_rank_order <= 3:
            total_lemmas += 1
            if model_rank_order[0] == 1:
                top_model_rank[0] += 1
            elif model_rank_order[0] == 2:
                top_model_rank[1] += 1
            elif model_rank_order[0] == 3:
                top_model_rank[2] += 1

            if user_rank_order == 1:
                top_user_rank[0] += 1
            elif user_rank_order == 2:
                top_user_rank[1] += 1
            elif user_rank_order == 3:
                top_user_rank[2] += 1

# Print statistics
print("Total Lemmas:", total_lemmas)
print("Top Gloss Ranked by Model:")
print("  Ranked 1st:", top_model_rank[0])
print("  Ranked 2nd:", top_model_rank[1])
print("  Ranked 3rd:", top_model_rank[2])
print("Top Gloss Ranked by User:")
print("  Ranked 1st:", top_user_rank[0])
print("  Ranked 2nd:", top_user_rank[1])
print("  Ranked 3rd:", top_user_rank[2])
```
